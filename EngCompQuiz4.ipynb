_In this quiz we will show that the logistic map (for certain growth rates) is a complex system. We will do this by showing evidence of evoluton, emergent behaviour and closeness to chaos._

_In class, the logistic map was introduced as a way to model the growth of a population of bacteria. We will use $x_n$ to denote the ratio between the population $p_n$ at time $n$ (hours) and its maximum carry capacity $M$. In other words:
  
  $x_n = P_n / M$  
  
We will also consider $r$ to be the growth rate of the bacteria, this will be a real number between 0 and 4. Then, the logistic map is given by the following iterated function:_  
  
| Equation # | Equation Expression |
|------------|---------------------|
| $(1)$ | $x_{n+1} = rx_n \cdot (1-x_n)$ |

_for $n$ = 1 , 2 , 3 , ...  
and an initial condition of $x_0$._

(a) Use the code `logistic_map.ipynb` (this is below the document for this Quiz on Cortex) to obtain a figure that shows the first 20 iterations of the solutions to equation (1)
with r=0.5 and different initial conditions  $x_0$=0.1 , $x_0$=0.3 , $x_0$=0.5 , $x_0$=0.7 , $x_0$=0.9
This should lead to a single figure with five solutions (one for each of the initial conditons)._

  **(a)**

```python
import numpy as np
import matplotlib.pyplot as plt 

#initialise r value
r = 0.5
#set number of iterations for the sequence
max_n = 20
#different initial conditions for the solutions
x0_vector = [0.1,0.3,0.5,0.7,0.9]
n = np.arange(1,max_n+1)
xn = np.zeros(max_n)

#for each of the different starting conditions, 
for x0 in x0_vector:
    #reinitialise x_0
    xn[0] = x0

    for i in np.arange(0,max_n-1):
        #equation #(1) per the quiz 4 question paper
        xn[i+1] = r * xn[i] * (1-xn[i])

    plt.plot(n, xn, 'o:', markersize=5, label='x_0='+str(x0)) 

# Adding major and minor ticks
plt.minorticks_on()

# Adding grid
plt.grid(True, which='major', linestyle='-', linewidth=0.8, alpha=0.8)  # Major grid
plt.grid(True, which='minor', linestyle='--', linewidth=0.5, alpha=0.5)  # Minor grid

plt.xticks(np.arange(0, max_n + 1, step=2))  # xticks every 2 hours
plt.yticks(np.arange(0, 1.1, step=0.1))     # yticks every 0.1 proportion

plt.title('Number of bacteria vs time for r='+str(r))
plt.ylabel('x_n (proportion of bacteria)')
plt.xlabel('n (hours)')
plt.legend()
```

* The equilibirum points occur when $x_{n+1} = x_n = x$
* This means there is no change between the current and next terms in the sequence. This is also represented by the change in population proportion with respect to time being zero.
#### Solutions by substitution:
* $x_{n+1} = rx_n \cdot (1-x_n)$  
* $x_{n+1} = x_n = x$  
  $x = rx(1-x)$  
  $x = rx -rx^2$  
  $0 = rx -rx^2 -x$  
  $0 = x(r -rx -1)$  
  $0 = x(r(1-x) -1)$  

* The two solutions for the generalised formula are:
1. $x = 0$
2. $r(1-x) -1 =0$  
   $x = 1 -1/r$

##### Plot (b) solutions:
